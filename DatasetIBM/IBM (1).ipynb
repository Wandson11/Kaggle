```python
#lista para guardar as estatisticas
stats = []

#descritiva por campo educacional
for campo in Dados['EducationField'].unique():
  temp = Dados[Dados['EducationField'] == campo]
  temp = temp[['EducationField', 'HourlyRate']]
  descriptive_temp = temp.describe()
  descriptive_temp.reset_index(inplace=True)
  descriptive_temp.rename(columns={'index': 'stats'}, inplace=True)
  stats.append(descriptive_temp)

#renomeia as colunas descritivas
for n in range(len(stats)):
  stats[n].rename(columns={'HourlyRate': Dados['EducationField'].unique()[n]}, inplace=True)


#gera dataframe com descritiva
stats_Dados = [Dados.set_index('stats') for Dados in stats]
stats_Dados[0].join(stats_Dados[1:])
```
