```python
def relatorio(dataframe, coluna):
  print(f'Quantidade de valores únicos: {dataframe[coluna].nunique()}')
  print(f'Quem são esses valores únicos: {dataframe[coluna].unique()}')
  print(f'Os valores únicos: {dataframe[coluna].isnull().sum()}')
  print(f'Como é a distribuição: \n{dataframe[coluna].value_counts()}')
```
