```python
import pandas as pd
import warnings

warnings.filterwarnings('ignore')
```

```python
treino = pd.read_csv('/content/train.csv')
teste = pd.read_csv('/content/test.csv')
```

```python
treino.head()
```

```python
treino.dropna(subset=['Survived'], inplace=True)
```

```python
y = treino['Survived']
```

```python
treino.drop(['Survived', 'PassengerId', 'Name', 'Ticket', 'Cabin'], inplace=True, axis=1)
```

```python
X = treino
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.7)
```

```python
colunas_numericas = [numeros for numeros in X_train.columns if X_train[numeros].dtype in ['int64', 'float64']]
```

```python
colunas_numericas
```

```python
colunas_categoricas = [categoricas for categoricas in X_train.columns if X_train[categoricas].dtype == 'object' and X_train[categoricas].nunique() < 10]
```

```python
colunas_categoricas
```

```python
todas_colunas = colunas_numericas + colunas_categoricas
X_train = X_train[todas_colunas].copy()
X_valid = X_valid[todas_colunas].copy()
X_teste = teste[todas_colunas].copy()
```

```python
X_train
```

```python
X_valid
```

```python
X_teste
```

```python
X_train.isnull().sum()
```

```python
#superada essa fase, hora das imputacoes e pre-processamento
```

```python
from sklearn.impute import SimpleImputer
```

```python
from sklearn.pipeline import Pipeline
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
from sklearn.compose import ColumnTransformer
```

```python
from xgboost import XGBRegressor
```

```python
#preprocessando dados numericos
imputando = SimpleImputer(strategy='median')

#pre-processando categoricos
categoricos = Pipeline(steps=[
    ('imp', SimpleImputer(strategy='most_frequent')),
    ('Onehot', OneHotEncoder(handle_unknown='ignore', sparse=False))
])

preprocessamento = ColumnTransformer(transformers=[
    ('num', imputando, colunas_numericas),
    ('cat', categoricos, colunas_categoricas)
])
```

```python
#escolhendo o modelo
modelo = XGBRegressor(n_estimators=250, learning_rate=0.1)
```

```python
#juntando tudo via pipeline
my_pipeline = Pipeline(steps=[
    ('prepr', preprocessamento),
    ('model', modelo)
])
```

```python
my_pipeline.fit(X_train, y_train)
```

```python
predicao = pd.DataFrame(my_pipeline.predict(X_valid))
predicao
```

```python
predicao[0].value_counts()
```

```python
#testando a validade dos dados via m.a.e.
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(y_valid, predicao)
```

```python
#validacao cruzada
from sklearn.model_selection import cross_val_score
```

```python
validacao = -1 * cross_val_score(my_pipeline, X, y, cv=5, scoring='neg_mean_absolute_error')
validacao
```

```python
validacao.mean()
```
