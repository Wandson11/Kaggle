# Função Resumo

```python
def space_funcao(dados, column, count=True):
  print(f'Quantidade de valores únicos: {dados[column].nunique()}')
  print(f'Quais são esses valores únicos: {dados[column].unique()}')
  print(f'Quantidade de valores nulos: {dados[column].isnull().sum()}')
  print(f'Quantidade de valores por categoria: {dados[column].value_counts()}')

  if count == True:
    sns.countplot(data = dados, x = column, hue='Transported')
  else:
    sns.displot(dados[column], hue='Transported')
```

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, label_binarize
from sklearn.model_selection import train_test_split
from sklearn import model_selection, tree, metrics, preprocessing, linear_model
from sklearn.svm import  LinearSVC
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LinearRegression, LogisticRegression, SGDClassifier
from sklearn.tree import DecisionTreeClassifier
!pip install catboost
import catboost
from catboost import CatBoostClassifier, Pool, cv
```

```python
def funcao_calculo(machine, X_train, y_train, vc):
  modelo = machine.fit(X_train, y_train)
  acuracia = round(modelo.score(X_train, y_train), 2)

  train_pred = model_selection.cross_val_predict(machine, X_train, y_train, cv = vc, n_jobs= -1)
  acuracia_vc = round(metrics.accuracy_score(y_train, train_pred)* 100, 2)

  return acuracia, acuracia_vc
```

```python
#random forest
acc_rf, acc_vc_rf = funcao_calculo(RandomForestClassifier(), X, y, 10)

print(f'Acurácia: {acc_rf}')
print(f'Acurácia validação cruzada: {acc_vc_rf}')
```

```python
#logistic regression
acc_lg, acc_vc_lg = funcao_calculo(LogisticRegression(), X, y, 10)

print(f'Acurácia: {acc_lg}')
print(f'Acurácia validação cruzada: {acc_vc_lg}')
```

```python
# k-Neurest Neighbours
acc_neuro, acc_vc_neuro = funcao_calculo(KNeighborsClassifier(), X, y, 10)

print(f'Acurácia: {acc_neuro}')
print(f'Acurácia validação cruzada: {acc_vc_neuro}')
```

```python
# Gaussian Naive Bayes

acc_gauss, acc_vc_gauss = funcao_calculo(GaussianNB(), X, y, 10)

print(f'Acurácia: {acc_gauss}')
print(f'Acurácia validação cruzada: {acc_vc_gauss}')
```

```python
# SVC Linear support vector machines

acc_svc, acc_vc_svc = funcao_calculo(LinearSVC( dual = False),X, y, 10)

print(f'Acurácia: {acc_svc}')
print(f'Acurácia validação cruzada: {acc_vc_svc}')
```

```python
# Stochatics gradient descent

acc_sgd, acc_vc_sgd = funcao_calculo(SGDClassifier(),X , y, 10)

print(f'Acurácia: {acc_sgd}')
print(f'Acurácia validação cruzada: {acc_vc_sgd}')
```

```python
#decision tree classifier

acc_tree, acc_vc_tree = funcao_calculo(DecisionTreeClassifier(), X, y, 10)

print(f'Acurácia: {acc_tree}')
print(f'Acurácia validação cruzada: {acc_vc_tree}')
```

```python
#gradient boost classifier

acc_gbc, acc_vc_gbc = funcao_calculo(GradientBoostingClassifier(), X, y, 10)

print(f'Acurácia: {acc_gbc}')
print(f'Acurácia validação cruzada: {acc_vc_gbc}')
```
